```python
import re
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
# from selenium_stealth import stealth
from bs4 import BeautifulSoup
import pandas as pd
import time
from tqdm.auto import tqdm
import numpy as np

driver = webdriver.Chrome('chromedriver', options=options)
```

```python
list_deskripsi_produk = []
list_kota = []
list_price = []
list_terjual = []
list_rating = []
list_nama = []
list_spek_color = []
list_penilaian = []
list_spesifikasi = []
list_deskripsi = []
list_link = []
link_error = []
list_notfound = []
```

# Main Code

```python
keyword = str(input('masukan input : '))
print(f'keyword {keyword} akan segera diproses untuk scraping')

link = f'https://shopee.co.id/search?keyword={keyword}'
```

```python
def get_data_link_shopee(link):
  driver.set_window_size(1920,1080)
  driver.get(link)
  time.sleep(60) # wait loading page

  data = BeautifulSoup(driver.page_source, 'html.parser')
  
  for a in tqdm(data.find_all('div', class_='col-xs-2-4 shopee-search-item-result__item')):
      try:
          base_link = 'https://shopee.co.id'
          link = a.find('a')['href']
          link = base_link + link
          list_link.append(link)
      except:
          
          list_notfound.append('')
  # print(f'the number of links that have been successfully scrapped : {len(list_link)}')
          
```

```python
def scraping_shopee(link):
  driver.set_window_size(1300, 800)
  driver.get(link)
  time.sleep(7)

  # scroll the page
  rentang = 500
  for i in range(1, 4):
    akhir = rentang * i
    perintah = "window.scrollTo(0,"+str(akhir)+")"
    driver.execute_script(perintah)
    # print('loading ke :', i)
    time.sleep(1.5)

  data = BeautifulSoup(driver.page_source, 'html.parser')

  try:
    penilaian = data.find_all('div', class_='flex IZIVH+')[1].text.replace('penilaian', '')
    list_penilaian.append(penilaian)
  except:
    list_penilaian.append('belum ada penilaian')

  try:
    spesifikasi = data.find_all('div', class_='U9rGd1')[0].text
    list_spesifikasi.append(spesifikasi)

    deskripsi = data.find_all('div', class_='U9rGd1')[1].text
    list_deskripsi.append(deskripsi)

  except:
    list_spesifikasi.append('')
    list_deskripsi.append('')

  for a in tqdm(data.find_all('div', class_='flex flex-auto RBf1cu')):
      name_product = a.find('span').text
      list_nama.append(name_product)
      try:
        rating = a.find('div', class_='_1k47d8 _046PXf').text
        list_rating.append(rating)
      
        terjual = a.find('div', class_='P3CdcB').text
        list_terjual.append(terjual)
      except:
        list_rating.append('')
        list_terjual.append('')
      
      price = a.find('div', class_='pqTWkA').text
      list_price.append(price)
      
      kota = a.find('span', class_='jqPI6Z').text
      list_kota.append(kota)
      

```

```python
for i in tqdm(range(0, 11)):
  link = f"https://shopee.co.id/search?keyword=laptop&page={i}"
  get_data_link_shopee(link)
print(f'the number of links that have been successfully scrapped : {len(list_link)}')
print(f'number of failed links fetched due to connection : {len(list_notfound)}')

  
```

```python
dictionary = {
    'link': list_link
}


df = pd.DataFrame.from_dict(dictionary, orient='index')
df = df.T
df = df['link']
df = df.drop_duplicates(keep='last')
# pd.options.display.max_rows = 10000
df
```

```python
i = 0
last = len(df)


for i in tqdm(range(i, last)):
  link = f'{df.iloc[i]}'
  if '【' in link:
    continue
  if '®' in link:
    continue
  scraping_shopee(link)
print('successful scraping')
```
